```python
import pandas as pd
import numpy as np
```

```python
gtfs_stops = pd.read_csv("gtfsv23/stops.txt",sep=",")
gtfs_trip = pd.read_csv("gtfsv23/trips.txt",sep=",")
gtfs_routes = pd.read_csv("gtfsv23/routes.txt",sep=",")
gtfs_shapes = pd.read_csv("gtfsv23/shapes.txt",sep=",")
#gtfs_services = pd.read_csv("gtfsv23/services.txt",sep=",")
gtfs_agency = pd.read_csv("gtfsv23/agency.txt",sep=",")
```

```python
gtfs_agency[:50]
route_agency = gtfs_routes.merge(gtfs_agency, on="agency_id")
route_agency =  route_agency.drop(['route_color','route_long_name','route_desc','route_url','agency_timezone','agency_url','route_text_color'],axis=1)
with pd.option_context('display.max_rows', None, 'display.max_columns',None):
    display(route_agency)
```

```python
rat = route_agency.merge(gtfs_trip , on = "route_id") # rat  = route agency trip merge table
with pd.option_context('display.max_rows', None, 'display.max_columns',None):
    display(rat[:1000])
```
